Importing the packages

```python
import numpy as np
import pandas as pd
from matplotlib  import pyplot as plt
import seaborn as sb
```

Reading the data

```python
diamonds=pd.read_csv("diamonds.csv")
diamonds
```

deleting the unwanted column

```python
diamonds2=diamonds.drop("Unnamed: 0",axis=1)
diamonds2.shape
```

To show  the information about dataset.

```python
diamonds2.info
```

To check if any null value or empty cells in the data.

```python
diamonds2.isnull().sum()
```

To see the  statisticall data means average and min and max values

```python
diamonds2.describe()
```

To delet the 0 contains rows 

```python
diamonds2.loc[(diamonds2['x']==0) | (diamonds2['y']==0) | (diamonds2['z']==0)]
```

```python
diamonds2 = diamonds2[(diamonds2[['x','y','z']] != 0).all(axis=1)]
```

To deleting the duplicate rows

```python
result=diamonds2.duplicated()
result.sum()
diamonds2[result]
```

```python
diamonds2.drop_duplicates(inplace=True)
```

```python
result=diamonds2.duplicated()
result.sum()
```
